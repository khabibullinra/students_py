# I. Расчет проведем по данным 1967 года; исходные данные берем из таблицы 3.10.2.1 и 3.10.2.2, используя промысловые данные:

## - пластовое давление	$Р_{пл} = 11,57 МПа$,
## - пластовая температура	$t_{пл} = 58 °С$,
## - давление насыщения	$Р_{нас} = 7,65 МПа$,
## - массовая газонасыщенность	$G_м = 65,56 \frac{м^3}{т}$, 
## - объемный коэффициент 	$b_н = 1,195$,
## - плотность пластовой нефти	$ρ_{нпл} = 773 \frac{кг}{м^3}$,
## - плотность дегазированной нефти	$ρ_{нд} = 854,1 \frac{кг}{м^3}$,
## - содержание метана	$у_м = 50,24$ %,
## - содержание азота	$у_a = 2,3$ %,
## - плотность газа	$ρ_г = 1,222 \frac{кг}{м^3}$.
## Расчет при $t = 20 °С$


# 1. Расчет функции $f_{ШБП}$:
## $f_{ШБП}=\frac{701,8}{G_м*(у_м-0,8*у_a)}$

```python
import math
%pylab inline
G_m3m3=65.56 
y_m=0.5024
y_a=0.023
P_nas_MPa=7.65
t_pl_C=58
ρnd_otn = 0.8541
ρg_otn = 0.945
f_шбп=math.trunc(100*701.8/(G_m3m3*(y_m-0.8*y_a)))/100 #Расчет функции fШБП 
print ('f_шбп=', f_шбп ) 
```

# 2. Расчет $Р_{нас20}$:
## $Р_{нас20} = Р_{нас} + \frac{20 – t_{пл}}{9,157 + f_{ШБП}}$

```python
Р_nas20_MPa=math.trunc(100*(P_nas_MPa +(20-t_pl_C)/(9.157+f_шбп)))/100 #Расчет давления насыщения при 20 
print ('Р_нас20=', Р_nas20_MPa, 'МПа' ) 
```

# 3. Задаем ряд значений $n_i$:
## $n_i=\frac{Р_i}{Р_{нас20}} =(1,0;0,8;0,6;0,4;0,2;0,1;\frac{0,1}{Р_{нас20}})$.
# 4. Вычисляем матрицу давлений $Р_i$:
## $Р_i = n_i · Р_{нас20}$

```python
import numpy as np
n_i = np.array([1, 0.8, 0.6, 0.4, 0.2, 0.1, 0.1/Р_nas20_MPa]) 
P_i_MPa=n_i*Р_nas20_MPa # Вычисляем матрицу давлений Рi 
print(P_i_MPa)
```

# 5. Рассчитываем коэффициенты $R_i$:
## $R_i=\frac{lg⁡(n_i)}{lg⁡(10∙Р_{нас20})}$ 

```python
def R_i(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000 #Рассчитываем коэффициенты  Ri
    return R_i
n_i = np.array([1, 0.8, 0.6, 0.4, 0.2, 0.1, 0.1/Р_nas20_MPa])
print ('R_i =', R_i(n_i)) 
```

# 6. Расчет коэффициентов $D_1 и U_1 (ρ ̅_{нд} = 0,8541; ρ ̅_г = 0,945)$:
## $D_1=4,06(ρ ̅_{нд}∙ρ ̅_г-1,045)$
## $U_1=ρ ̅_{нд}∙G_м-186$

```python
D_1= math.trunc(100*4.06* (ρnd_otn * ρg_otn-1.045))/100 #Расчет коэффициентов  D1иU1
U_1= math.trunc(100*(ρnd_otn*G_m3m3-186))/100
print ('D_1=', D_1)
print ('U_1=', U_1)
```

# 7. Расчет объема выделяющегося из нефти газа:
# $G_{м,i}=Г_м∙R_i [D_1 (1+R_i )-1]$

```python
def G_i_m3tonna(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    return np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100   #Расчет объема выделяющегося из нефти газа                                  
print ('G_m,i =', G_i_m3tonna(n_i))
x = P_i_MPa
y = G_i_m3tonna(n_i)
plot(x, y)
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Объем выделяющегося из нефти газа, м^3/тонна')
show()                           
```

# 8. Расчет газонасыщенности пластовой нефти $Г_{м,i}$:
## $Г_{м,i}=Г_м-G_{м,i}$

```python
def Г_i_m3tonna(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    return G_m3m3-G_i_m3tonna #Расчет газонасыщенности пластовой нефти  
print ('Г_i =', Г_i_m3tonna(n_i))
x = P_i_MPa
y = Г_i_m3tonna(n_i)
plot(x, y)
plt.title('Г_i=f(P)')
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Газонасыщенность пластовой нефти, м^3/тонна')
show()
```

# 9. Расчет относительной плотности выделяющегося из нефти газа:
## $ρ ̅_{гвi}=ρ ̅_г-0,0036(1+R_i)(105,7+U_1∙R_i)$

```python
def ρ_gvi_otn(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    return np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000 #Расчет относительной плотности выделяющегося из нефти газа
print ('ρ_гвi =', ρ_gvi_otn(n_i))
x = P_i_MPa
y = ρ_gvi_otn(n_i)
plot(x, y)
plt.title('ρ_гвi=f(P)')
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Относительная плотность выделяющегося из нефти газа')
show()
```

# 10. Расчет относительной плотности газа, остающегося в нефти в растворенном состоянии:
## $ρ ̅_{гр,i}=\frac{Г_м}{Г_{м,i}}*(ρ ̅_г-ρ ̅_{гв,i}*\frac{G_{м,i}}{Г_м})$

```python
def ρ_gri_otn(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    Г_i_m3tonna=G_m3m3-G_i_m3tonna
    ρ_gvi_otn=np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000
    return np.trunc(1000*G_m3m3/Г_i_m3tonna*(ρg_otn-ρ_gvi_otn*G_i_m3tonna/G_m3m3))/1000 #Расчет относительной плотности газа, 
                                                                                  # остающегося в нефти в растворенном состоянии
print ('ρ_грi =', ρ_gri_otn(n_i))   
```

## $ρ ̅_{гр,7}$ должна быть найдена экстраполяцией графической зависимости $ρ ̅_{гр,i}=f(Р)$ для Р = 0,1 МПа.

```python
from pylab import *
import numpy as np
x=[0.643, 1.286,  2.572,  3.858,  5.144, 6.43]
y=[2.022, 1.492,1.181,1.056,0.988,0.945,]
z = np.polyfit(x, y, 4)
f = np.poly1d(z)
x_new = np.linspace(0, 6.43, 50)
y_new = f(x_new)
plot(x_new, y_new)
plt.title('ρ_грi=f(P)')
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Относительная плотность газа, оставшегося в нефти')
show()
x_1=0.1
y_1=np.trunc(100*f(0.1))/100
print ('ρ_гр7=', y_1)
```

```python
def ρ_gri_otn_new(n_i):
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    Г_i_m3tonna=G_m3m3-G_i_m3tonna
    ρ_gvi_otn=np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000
    ρ_gri_otn=np.trunc(1000*G_m3m3/Г_i_m3tonna*(ρg_otn-ρ_gvi_otn*G_i_m3tonna/G_m3m3))/1000
    ρ_gri_otn_new=np.append(ρ_gri_otn[:6],y_1) #подставил в массив относительной плотности газа, значение найденное экстраполяцией
    return ρ_gri_otn_new
print(ρ_gri_otn_new(n_i))
```

## 11. Расчет безразмерного параметра $λ_i$:
## $λ_i=3,54(1,2147-ρ ̅_{нд} )+1,0337ρ ̅_{грi}+5,581ρ ̅_{нд} (1-1,61ρ ̅_{нд}∙10^{-3} Г_{мi})10^{-3} Г_{мi}$

```python
def λ_i(n_i): #Расчет безразмерного параметра  
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    Г_i_m3tonna=G_m3m3-G_i_m3tonna
    ρ_gvi_otn=np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000
    ρ_gri_otn=np.trunc(1000*G_m3m3/Г_i_m3tonna*(ρg_otn-ρ_gvi_otn*G_i_m3tonna/G_m3m3))/1000
    ρ_gri_otn_new=np.append(ρ_gri_otn[:6],y_1)
    return np.trunc(1000*(3.54*(1.2147-ρnd_otn)+1.0337*ρ_gri_otn_new+5.581*ρnd_otn*(1-1.61*ρnd_otn/1000*Г_i_m3tonna)*Г_i_m3tonna/1000))/1000           
print (' λ_i=', λ_i(n_i))
```

# 12. Расчет объемного коэффициента нефти:
## $b_i=1+1,0733ρ ̅_{нд}∙λ_i∙10^{-3}∙Г_{мi}-6,5∙10^{-4}∙Р_i$

```python
def b_i_m3m3(n_i): #Расчет объемного коэффициента нефти
    P_i_MPa=n_i*Р_nas20_MPa
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    Г_i_m3tonna=G_m3m3-G_i_m3tonna
    ρ_gvi_otn=np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000
    ρ_gri_otn=np.trunc(1000*G_m3m3/Г_i_m3tonna*(ρg_otn-ρ_gvi_otn*G_i_m3tonna/G_m3m3))/1000
    ρ_gri_otn_new=np.append(ρ_gri_otn[:6],y_1)
    λ_i=np.trunc(1000*(3.54*(1.2147-ρnd_otn)+1.0337*ρ_gri_otn_new+5.581*ρnd_otn*(1-1.61*ρnd_otn/1000*Г_i_m3tonna)*Г_i_m3tonna/1000))/1000
    return np.trunc((1+1.0733*ρnd_otn*λ_i/1000*Г_i_m3tonna-6.5/10000*P_i_MPa)*1000)/1000     
print (' b_i=', b_i_m3m3(n_i))
x = P_i_MPa
y = b_i_m3m3(n_i)
plot(x, y)
plt.title('b_i=f(P)')
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Объемный коэффициент нефти, м^3/м^3')
show()
```

# 13. Расчет относительной плотности газонасыщенной нефти:
## $ρ ̅_{нг,i}=\frac{ρ ̅_{нд}}{b_{нi}}(1+1,293ρ ̅_{грi}∙10^{-3}∙Г_{мi} )$

```python
def ρ_ngi_otn(n_i): #Расчет относительной плотности газонасыщенной нефти
    P_i_MPa=n_i*Р_nas20_MPa
    R_i=np.trunc(1000*np.log10(n_i)/np.log10(10*Р_nas20_MPa))/1000
    G_i_m3tonna=np.trunc(100*G_m3m3*R_i*(D_1*(1+R_i)-1))/100 
    Г_i_m3tonna=G_m3m3-G_i_m3tonna
    ρ_gvi_otn=np.trunc(1000*(ρg_otn-0.0036*(1+R_i)*(105.7+U_1*R_i)))/1000
    ρ_gri_otn=np.trunc(1000*G_m3m3/Г_i_m3tonna*(ρg_otn-ρ_gvi_otn*G_i_m3tonna/G_m3m3))/1000
    ρ_gri_otn_new=np.append(ρ_gri_otn[:6],y_1)
    λ_i=np.trunc(1000*(3.54*(1.2147-ρnd_otn)+1.0337*ρ_gri_otn_new+5.581*ρnd_otn*(1-1.61*ρnd_otn/1000*Г_i_m3tonna)*Г_i_m3tonna/1000))/1000
    b_i_m3m3=np.trunc((1+1.0733*ρnd_otn*λ_i/1000*Г_i_m3tonna-6.5/10000*P_i_MPa)*1000)/1000
    return np.trunc(1000*ρnd_otn/b_i_m3m3*(1+1.293*ρ_gri_otn_new/1000*Г_i_m3tonna))/1000
print (' ρ_нгi=', ρ_ngi_otn(n_i))
x = P_i_MPa
y = ρ_ngi_otn(n_i)
plot(x, y)
plt.title('ρ_нгi=f(P)')
plt.xlabel(u'Давление, МПа')
plt.ylabel(u'Относительная плотность газонасыщенной нефти')
show()
```
