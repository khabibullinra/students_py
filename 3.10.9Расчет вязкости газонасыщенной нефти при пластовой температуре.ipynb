# Расчет вязкости газонасыщенной нефти $μ_{нг}$ при пластовой температуре, а также при изменении газонасыщенности нефти и вязкости дегазированной нефти при $t_{пл}$ и атмосферном давлении базируется на формуле Чью и Коннели:
## $μ ̅_{нг}=А∙μ ̅_{ндt}^B$, 
# где $μ ̅_{нг}$ – относительная вязкость газонасыщенной нефти при температуре t и давлении насыщения;
# $μ ̅_{ндt}$ – относительная вязкость дегазированной нефти при температуре t и атмосферном давлении;
# А и В – коэффициенты, представленные в виде графика и зависящие от количества растворенного в нефти газа (газонасыщенности), аппроксимированные проф. Мищенко И.Т. в аналитическом виде:
## $А = exp [-87,24·10^{-4} G_o^*+12,9·10^{-6} (G_o^*)^2]$,
## $В = exp [-47,11·10^{-4} G_o^*+8,3·10^{-6} (G_o^*)^2]$, 
# $G_o^*$ – газонасыщенность нефти (объем газа приведен к 15 °С и атмосферному давлению, $\frac{м^3}{м^3}$)
## $G_o^* = 0,983 (1+α_н·5) G_o$,
# $G_o$ – газонасыщенность нефти (объем газа приведен к 20 °С и атмосферному давлению, $\frac{м^3}{м^3}$),
# $α_н$ – коэффициент термического расширения дегазированной нефти:
## $α_н=10^{-3}*2,638(1,169-ρ ̅_{нд})$               $при 0,78≤ρ ̅_{нд}≤0,86$
## $α_н=10^{-3}*1,975(1,272-ρ ̅_{нд} )$              $при 0,86<ρ ̅_{нд}≤0,96$
# где $ρ ̅_{нд}$ – относительная плотность дегазированной нефти.
# Рассчитаем вязкость газонасыщенной нефти при пластовой температуре и давлении насыщения.



```python
import math
t_пл = 57
μ_нд60  = 12.21
Go = 56.4 
ρ_нд = 0.8541 
if 0.78 <= ρ_нд < 0.86: α_н=2.638/1000*(1.169-ρ_нд)
elif  0.86 <= ρ_нд <= 0.96: α_н=1.975/1000*(1.272-ρ_нд)
else: print('Неправильное значение плотности!')
print ('α_н=',math.trunc(α_н*100000)/100000  )     
Go1 = 0.983*(1+α_н*5)*Go
print ('Go*=',math.trunc((Go1)*10)/10, 'м^3/м^3'  )
A=math.exp(-87.24/10000*Go1+12.9/1000000*pow(Go1,2))
B=math.exp(-47.11/10000*Go1+8.3/1000000*pow(Go1,2))
print ('A=',math.trunc(A*1000)/1000  )
print ('B=',math.trunc(B*1000)/1000  )
μ_нг=A*pow(μ_нд60,B)
print ('μ_нг=',math.trunc(μ_нг*100)/100, 'мПа*с'  )
```
